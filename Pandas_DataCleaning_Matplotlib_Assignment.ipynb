## Q1. Import pandas and matplotlib.pyplot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df.duplicated().sum()
```

## Q11. Remove duplicate rows

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

## Q12. Remove extra spaces from Title

```python
df['Title'] = df['Title'].str.strip()
df['Title']
```

## Q13. Convert Title to Proper Case

```python
df['Title'] = df['Title'].str.title()
df['Title']
```

## Q14. Correct Genre spellings

```python
df['Genre'] = df['Genre'].replace({'Acton':'Action','Scifi':'Sci-Fi','Romnce':'Romance'})
df['Genre'].unique()
```

## Q15. Convert Country to upper case

```python
df['Country'] = df['Country'].str.upper()
df['Country']
```

## Q16. Convert Language to lower case

```python
df['Language'] = df['Language'].str.lower()
df['Language']
```

## Q17. Convert Rating to float

```python
df['Rating'] = df['Rating'].astype(float)
df['Rating']
```

## Q18. Clean Duration column

```python
df['Duration'] = df['Duration'].str.replace('min','').astype(int)
df['Duration']
```

## Q19. Clean Budget column

```python
df['Budget'] = df['Budget'].str.replace('$', '', regex=False).str.replace('M', '', regex=False).astype(float)
df['Budget']
```

## Q20. Clean Revenue column

```python
df['Revenue'] = df['Revenue'].str.replace(',','').astype(float)
df['Revenue']
```

## Q21. Convert Release_Year to integer

```python
df['Release_Year'] = df['Release_Year'].astype(int)
df['Release_Year']
```

## Q22. Convert Added_Date to datetime

```python
df['Added_Date'] = pd.to_datetime(df['Added_Date'], format='%d-%m-%Y')
df['Added_Date']
```

## Q23. Delete rows where Title is missing

```python
df['Title'].isnull().sum()
df.dropna(subset=['Title'], inplace=True)
df['Title'].isnull().sum()
df['Title']

#Language
df['Language'].isnull().sum()
df.dropna(subset=['Language'], inplace=True)
df['Language'].isnull().sum()
df['Language']
```

## Q24. Fill missing Rating with mean

```python
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
df['Rating']
```

## Q25. Fill missing Country with mode

```python
df['Country'].fillna(df['Country'].mode()[0], inplace=True)
df['Country']
```

## Q26. Fill missing Budget with median

```python
df['Budget'].fillna(df['Budget'].median(), inplace=True)
df['Budget']

#Revenue
df['Revenue'].fillna(df['Revenue'].median(), inplace=True)
df['Revenue']
```

## Q27. Check remaining null values

```python
df.isnull().sum()
```

## Q28. Save cleaned dataset

```python
df
```

## Q30. Average rating by platform

```python
df.groupby('Platform')['Rating'].mean()
```

## Q31. Bar chart of platform-wise movie count

```python
plt.bar(df['Platform'].value_counts().index, df['Platform'].value_counts().values)
plt.xlabel('Platform')
plt.ylabel('Movie Count')
plt.title('Platform-wise Movie Count')
plt.show()
```

## Q32. Pie chart of genre distribution

```python
plt.pie(df['Genre'].value_counts(), labels=df['Genre'].value_counts().index, autopct='%1.1f%%')
plt.title('Genre Distribution')
plt.show()
```

## Q33. Histogram of Ratings

```python
plt.hist(df['Rating'], bins=10)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Rating Distribution')
plt.show()
```

## Q34. Line chart of release year counts

```python
plt.plot(df['Release_Year'].value_counts().sort_index().index, df['Release_Year'].value_counts().sort_index().values)
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Release Year Counts')
```

## Q35. Box plot of Ratings

```python
plt.boxplot(df['Rating'])
plt.ylabel('Rating')
plt.title('Rating Distribution')
plt.show()
```

## Q36. Scatter plot of Budget vs Revenue

```python
plt.scatter(df['Budget'], df['Revenue'])
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget vs Revenue')
plt.show()
```
