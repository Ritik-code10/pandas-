# Q1. Select Single Column

## Problem Statement
Create a DataFrame and display only the Name column.

## Sample Input
```python
Name, Age
```

## Sample Output
```python
Only Name column
```


````python
# Write your code here
import pandas as pd
df = pd.DataFrame({
    'Name' : ['Aman','Raj','Ritik','Amit'],
    'Age' : [20,21,23,22],
    'Marks' : [90,80,70,60]
})
print(df['Name'])
````


# Q2. Select Multiple Columns

## Problem Statement
Display Name and Marks columns from a DataFrame.

## Sample Input
```python
Name, Marks
```

## Sample Output
```python
Selected columns
```


````python
# Write your code here
print(df[['Name','Marks']])
````


# Q3. Filter Students

## Problem Statement
Display students with Marks greater than 80.

## Sample Input
```python
Marks > 80
```

## Sample Output
```python
Filtered rows
```


````python
# Write your code here
print(df[df['Marks']>80])
````


# Q4. Filter by City

## Problem Statement
Display students belonging to Bhopal.

## Sample Input
```python
City = Bhopal
```

## Sample Output
```python
Matching rows
```


````python
# Write your code here
df['City'] = ['Bhopal','Indore','Bhopal','Indore']
print(df[df['City']=='Bhopal'])
````


# Q5. Find Maximum Marks

## Problem Statement

## Sample Input

```python
Marks column
```

## Sample Output

```python
Highest value
```


# Q7. Sort Data

Sort students by Marks.

```python
Marks column
```


# Q8. Value Counts

## Problem Statement
Count occurrences of each city.

## Sample Input
```python
City column
```

## Sample Output
```python

```


````python
# Write your code here
print(df['City'].value_counts())
````


# Q9. Group By City

## Problem Statement
Find average marks city-wise.

## Sample Input
```python
City and Marks
```

## Sample Output
```python
Grouped result
```


````python
# Write your code here
print(df.groupby('City')['Marks'].mean())
````


# Q10. Multiple Conditions

## Problem Statement
Display students with Marks > 70 and Age > 20.

## Sample Input
```python
Marks and Age
```

## Sample Output
```python
Filtered rows
```


````python
# Write your code here
print(df[(df['Marks']>70) & (df['Age']>20)])
````
