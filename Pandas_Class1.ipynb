## 3 Understanding Series & DataFrame
- **Series**: One-dimensional labeled array, like a column in Excel.
- **DataFrame**: Two-dimensional table, similar to an Excel sheet or SQL table.

### Creating a Series from Python Lists or Dictionaries

````python
import pandas as pd
marks = pd.Series([30,40,50,60,70])
marks
marks[2]  #indexing
````

````python
import pandas as pd
marks = pd.Series([30,40,50,60,70],index =['Ronaldo','Messi','Virat','Dhoni','Mbappe'])
marks
# marks['Messi']
print(marks)
````

### Creating a DataFrame

````python
#Multiple columns -> DataFrame
info = pd.DataFrame(
    {
    'Name' : ['Rahul','Devraj','Abhishek','Rishi'],
    'age' : [20,21,22,23],
    'marks' : [100,80,90,85]
    })
print(info)
#selecting a column
print(info['Name'])

#To get specific row we use .loc
print([info.loc[1]])

#select a specific Value
print(info.loc[1,'Name'])
````

### Creating DataFrame from NumPy Arrays

````python
information = pd.DataFrame([['Rahul',52,80],['Devraj',22,85],['Abhishek',60,55]],columns = ['Name','Age','Marks']
                           )
print(information)
print(information.loc[2])
print(information.loc[2,'Marks'])
````

````python
#Generate a 3X4 matrix with random numbers (1-13) using numpy
#name the column as [A,B,C,D]
import numpy as np
import pandas as pd
arr = np.random.randint(1,100,size = (3,4))
arr_df = pd.DataFrame(arr,columns = ['A','B','C','D'])
print(arr_df)
print(arr_df.loc[1])
print(arr_df.loc[1,'B'])
````
