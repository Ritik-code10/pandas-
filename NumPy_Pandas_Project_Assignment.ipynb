## Question 15

Create another DataFrame containing **EmpID** and **Experience** and merge it with the Employee DataFrame.

Second DataFrame:

| EmpID | Experience |
|------:|-----------:|
|101|2|
|102|5|
|103|3|
|104|6|
|105|4|

Expected Output: Final merged DataFrame with Experience column.

````python
# Write your code here
df = pd.DataFrame({
    "EmpID":[101,102,103,104,105],
    "Name":["Raj","Amit","Neha","Priya","Rohan"],
    "Department":["IT","HR","IT","Finance","HR"],
    "Salary":[50000,42000,65000,58000,45000],
    "Age":[24,28,26,30,25]
})
df2 = pd.DataFrame({
    "EmpID":[101,102,103,104,105],
    "Experience":[2,5,3,6,4]
})
final_dataframe = pd.merge(df,df2,on='EmpID')
final_dataframe
````
