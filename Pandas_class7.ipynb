## 1. Using `concat()` for Stacking Data
- `pd.concat()` combines DataFrames vertically (rows) or horizontally (columns).
- Use `axis=0` for vertical stacking (default), `axis=1` for horizontal.
- `ignore_index=True` can reset the index after concatenation.

```python
import pandas as pd
# Dataset 1: Product details
df1 = pd.DataFrame({
    "ProductID" : [101,102,103],
    "ProductName" : ['Pen','Notebook','Marker'],
    "Price" : [10,50,30]
})

# Dataset 2: More products (same columns as df1 → good for vertical stacking)
df2 = pd.DataFrame({
    'ProductID': [104, 105],
    'ProductName': ['Pencil', 'Eraser'],
    'Price': [5,4]
})
# Dataset 3: Additional info (different columns → good for horizontal stacking)
df3 = pd.DataFrame({
    'Stock': [100, 60, 80],
    'Rating': [4.5,4.7,4.8]
})

print(f'DF1 {df1}')
print(f'DF2 {df2}')
print(f'DF3 {df3}')


#1. Vertical Stacking -> By Default verical stacking in rows wies (axis=0)
vertical_stacking = pd.concat([df1,df2],ignore_index=True)
vertical_stacking

#2.Horizontal Stacking (axis=1)
horizontal_stacking = pd.concat([df1,df3],axis=1)
horizontal_stacking
```

## 2. Using `merge()` for SQL-Style Joins
- `merge()` is similar to SQL joins.
- Specify `on` or `left_on`/`right_on` to choose key columns.
- Join types: `inner` (default), `left`, `right`, `outer`.

```python
# You guys have to concat df1 and df2 row wise
product = pd.concat([df1,df2],ignore_index=True)
product

sales = pd.DataFrame({
    "ProductID" : [101,102,104,106],
    "UnitsSold" : [500,200,150,50]
})
print(product)
print(sales)


#1. Inner join (Dono DataFrames ke beech only matching productID)
#          left   right
inner = pd.merge(product,sales,how='inner')
inner

#2. left join
left = pd.merge(product,sales,how='left')
left

#3. Right join

right = pd.merge(product,sales,how='right')
right

#4. Outer join (Everything, even unmatched)
outer = pd.merge(product,sales,how='outer')
outer


```

## 3. Using `join()` with Index Alignment
- `join()` joins DataFrames on their index or on a key column.
- By default, joins on index; can specify `on='col'`.
- Useful for adding columns to an existing DataFrame.

```python
sales = pd.DataFrame({
    "UnitsSold" : [500,200,150,50,350]
},index=[101,102,103,104,105])
print(product)
print(sales)

#1. join stock info to products
p_indexed = product.set_index('ProductID')
joined = p_indexed.join(sales)
joined
```
