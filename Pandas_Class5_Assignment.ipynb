## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
# Write your code here
import pandas as pd
Age = [25, 18, 30, 22]
df = pd.DataFrame(Age,columns=['Age'])
df.sort_values(by='Age')
```

## Q2. Sort Marks in Descending Order

Input:
Marks = [70, 95, 80, 60]

Output:
95
80
70
60

```python
# Write your code here
Marks = [70,95,80,60]
df = pd.DataFrame(Marks,columns=['Marks'])
print(df.sort_values(by='Marks',ascending=False))
```

## Q8. Count City Frequency

Input:
City = ['Bhopal','Indore','Bhopal','Jabalpur']

Output:
Bhopal      2

```python
# Write your code here
df = pd.DataFrame({"City" : ["Bhopal","Indore","Bhopal","Jabalpur"]})
print(df.value_counts())
```

## Q9. Find Percentage Frequency

Input:
City = ['A','A','B','C']

Output:
A = 50%
B = 25%
C = 25%

```python
# Write your code here
df = pd.DataFrame({"City" : ['A','A','B','C']})
print(df.value_counts(normalize=True)*100)
```

## Q14. Student Statistics

Input:
Marks = [60,70,80,90]

Output:
Mean = 75
Median = 75
Max = 90
Min = 60

```python
# Write your code here
df = pd.DataFrame({'Marks' : [60,70,80,90]})
print(f"Mean = {df.mean()}")
print(f"Median = {df.median()}")
print(f"Max = {df.max()}")
print(f"Min = {df.min()}")
```

## Q15. Mini Practice

Create a DataFrame with:
Name, Age, Score, Height

Tasks:
1. Sort by Score (Descending)
2. Find Mean Age
3. Find Median Height
4. Show Score Frequency

```python
# Write your code here
df = pd.DataFrame({"Name" : ['A','B','C','D'],
                   "Age" : [10,20,30,40],
                   "Score" : [60,80,70,90],
                   "Height" : [150,160,170,180]})
print(df.sort_values(by='Score',ascending=False))
print(df.mean(numeric_only=True))
print(df.median(numeric_only=True))
print(df.value_counts())
```
