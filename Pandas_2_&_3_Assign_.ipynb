# Q1. Display First Rows

## Problem Statement
Create the following DataFrame:

| Name | Age |
|------|-----|
| Aman | 21 |
| Rahul | 22 |
| Neha | 20 |
| Priya | 23 |
| Ravi | 24 |

Display the first 3 rows using `head()`.

## Sample Output
```python
First 3 rows of DataFrame
```


````python
# Write your code here
import pandas as pd
df = pd.DataFrame({"Name" : ['Aman','Rahul','Neha','Priya','Ravi'],
                   "Age" : [21,22,20,23,24]})
print(df.head(3))
````


# Q6. Select Data using iloc

## Problem Statement
Create a DataFrame with at least 4 rows.

Using `iloc`, print:
- Second row
- First two columns

## Sample Output
```python
Selected row values
```


````python
# Write your code here
df = pd.DataFrame({"Name":['Aman','Rahul','Neha','Priya'],
                   "Age" : [21,23,22,24],
                   'City': ['Bhopal','Indore','Bhopal','Indore']})
print(df.iloc[1])
print(df.iloc[0:2,0:2])
````


# Q7. Rename a Column

## Problem Statement
Create a DataFrame:

```python
Name
Marks
```

Rename column `Marks` to `Score`.

## Sample Output
```python
Name  Score
```


````python
# Write your code here
df = pd.DataFrame({"Name" : ['Aman','Raj','Amit','Ritik'],
                   "Marks" : [75,67,96,80]})
print(df.rename(columns={'Marks':'Score'}))
````


# Q8. Handle Missing Values

## Problem Statement

```python
Name = ["Aman", None, "Neha"]
Marks = [80, 75, None]
```

Fill missing values using `fillna()`.

## Sample Output

```python
Missing values replaced
```


# Q9. Replace Values

## Problem Statement
Create a DataFrame:

```python
City = ["BPL", "IND", "BPL"]
```

Replace:
- BPL → Bhopal
- IND → Indore

## Sample Output
```python
Bhopal
Indore
Bhopal
```


````python
# Write your code here
df = pd.DataFrame({"City" : ["BPL", "IND", "BPL"]})
print(df.replace({'BPL':'Bhopal','IND':'Indore'}))
````


# Q10. Remove Duplicate Records

## Problem Statement
Create a DataFrame:

```python

```

## Sample Output
