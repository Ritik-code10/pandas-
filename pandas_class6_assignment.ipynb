# Q1. Display Unique Values

## Problem Statement
Create a DataFrame with a City column and display all unique cities.

## Sample Input
```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output
```python
['Bhopal' 'Indore' 'Delhi']
```


````python
# Write your code here
import pandas as pd
df = pd.DataFrame({"City" : ["Bhopal", "Indore", "Bhopal", "Delhi"]})
print(df['City'].unique())
````


# Q4. Group By City

## Problem Statement
Find average marks city-wise using groupby().

## Sample Input
```python
City   Marks
Bhopal 80
Bhopal 90
Delhi  70
```

## Sample Output
```python
Bhopal 85
Delhi  70
```


````python
# Write your code here
df = pd.DataFrame({"City" : ["Bhopal", "Bhopal", "Delhi"], "Marks" : [80, 90, 70]})
avg_marks = df.groupby('City')['Marks'].mean()
print(avg_marks)
````


# Q5. Sort Values

## Problem Statement

````python
# Write your code here
df = pd.DataFrame({"Student" : ["Aman","Raj","Amit","Aryan"],
                   "Marks" : [80,50,60,89]})
print(df.sort_values("Marks"))
````


# Q8. Multiple Conditions

## Problem Statement
Display students whose marks are greater than 70 and age is greater than 20.

## Sample Output
```python
Filtered rows
```


````python
# Write your code here
df = pd.DataFrame({"Student" : ["Aman","Raj","Amit","Aryan"],
                   "Marks" : [80,50,60,89],
                   "Age" : [18,19,20,21]})
print(df[(df['Marks']>70)&(df['Age']>20) ])
````


# Q9. Group By Department

## Problem Statement
Create Employee and Department data and find total salary department-wise.

## Sample Output
```python
HR       50000
IT      120000
```


````python
# Write your code here
df = pd.DataFrame({
    "Employee": ["John", "Sarah", "Mike", "Anna", "Tom", "Laura", "Steve", "Kate"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 70000, 72000, 65000, 60000, 52000, 75000, 58000]
})
total_salary = df.groupby('Department')['Salary'].sum()
print(total_salary)
````


# Q10. Complete Practice Question

## Problem Statement
Create a DataFrame containing:
- Name
- Age
- City
- Marks

Perform:
1. Show unique cities
2. Count city frequency
3. Filter marks > 80
4. Sort by marks

## Sample Output
```python
Required Results
```


````python
# Write your code here
df = pd.DataFrame({"Name" : ['Ritk','Raj','Aryan','Abhishak'],
                   "Age" : [20,23,21,25],
                   "City" : ['Bhopal','Indore','Bhopal','Ujjain'],
                   "Marks" : [80,77,69,87]})
print(df['City'].unique())
print(df['City'].value_counts())
print(df[df['Marks']>80])
print(df.sort_values('Marks'))
````
